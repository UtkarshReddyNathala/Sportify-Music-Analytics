```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

core_df = pd.read_csv(r"D:\spotify\data\raw\core\spotify_data clean.csv")
audio_df = pd.read_csv(r"D:\spotify\data\raw\core\SpotifyAudioFeaturesApril2019.csv")
```

```python
def clean_columns(df):
    df.columns = (
        df.columns
        .str.lower()
        .str.strip()
        .str.replace(" ", "_")
        .str.replace(r"[^\w_]", "", regex=True)
    )
    return df

core_df = clean_columns(core_df)
audio_df = clean_columns(audio_df)
```

```python
core_df = core_df.loc[:, ~core_df.columns.str.contains("unnamed")]
```

```python
audio_df = audio_df.loc[:, ~audio_df.columns.str.contains("unnamed")]
```

```python
core_df.duplicated().sum()
audio_df.duplicated().sum()
```

```python
core_df = core_df.drop_duplicates()
audio_df = audio_df.drop_duplicates()
```

```python
if "popularity" in core_df.columns:
    core_df["popularity"] = pd.to_numeric(core_df["popularity"], errors="coerce")
```

```python
if "year" in core_df.columns:
    core_df["year"] = pd.to_numeric(core_df["year"], errors="coerce")
```

```python
critical_cols = ["track_name", "artist_name"]

core_df = core_df.dropna(subset=[c for c in critical_cols if c in core_df.columns])
audio_df = audio_df.dropna()
```

```python
core_df["merge_key"] = (
    core_df["track_name"].str.lower().str.strip() + "_" +
    core_df["artist_name"].str.lower().str.strip()
)

audio_df["merge_key"] = (
    audio_df["track_name"].str.lower().str.strip() + "_" +
    audio_df["artist_name"].str.lower().str.strip()
)
```

```python
charts_df = pd.read_csv(r"D:\spotify\data\raw\supporting\spotify_top_songs_audio_features.csv")
charts_df = clean_columns(charts_df)
charts_df = charts_df.drop_duplicates()
```

```python
genre_df = pd.read_csv(r"D:\spotify\data\raw\supporting\dataset.csv")
genre_df = clean_columns(genre_df)
genre_df = genre_df.loc[:, ~genre_df.columns.str.contains("unnamed")]
```

```python
core_df.to_csv(r"D:\spotify\data\processed\cleaned\core_cleaned.csv", index=False)
audio_df.to_csv(r"D:\spotify\data\processed\cleaned\audio_features_cleaned.csv", index=False)
charts_df.to_csv(r"D:\spotify\data\processed\cleaned\top_charts_cleaned.csv", index=False)
genre_df.to_csv(r"D:\spotify\data\processed\cleaned\genres_cleaned.csv", index=False)
```
