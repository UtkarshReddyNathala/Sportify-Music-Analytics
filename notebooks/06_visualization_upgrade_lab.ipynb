```python
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go

pd.set_option("display.max_columns", None)
```

```python
yearly_df = pd.read_csv(r"D:\spotify\data\processed\analytics_ready\yearly_trends.csv")
```

```python
yearly_df.head()
```

```python
features = ["danceability", "energy", "valence"]

fig = px.line(
    yearly_df,
    x="year",
    y=features,
    markers=True,
    title="Evolution of Key Spotify Audio Features Over Time"
)

fig.update_layout(
    xaxis_title="Year",
    yaxis_title="Feature Value",
    legend_title="Audio Feature",
    template="plotly_dark"
)

fig.show()
```

```python
fig = px.line(
    yearly_df,
    x="year",
    y="danceability",
    markers=True,
    title="Danceability Trend with Pandemic Period Highlight"
)

fig.add_vrect(
    x0=2020, x1=2021,
    fillcolor="red",
    opacity=0.15,
    annotation_text="Pandemic Era",
    annotation_position="top left"
)

fig.update_layout(template="plotly_dark")
fig.show()
```

```python
genre_df = pd.read_csv(r"D:\spotify\data\processed\analytics_ready\genre_audio_analysis.csv")
```

```python
genre_df.head()
```

```python
genre_counts = (
    genre_df["track_genre"]
    .value_counts()
    .reset_index()
)

genre_counts.columns = ["track_genre", "count"]

fig = px.bar(
    genre_counts,
    x="track_genre",
    y="count",
    title="Distribution of Tracks Across Genres",
    color="count",
    template="plotly_dark"
)

fig.update_layout(
    xaxis_title="Genre",
    yaxis_title="Number of Tracks"
)

fig.show()
```

```python
feature = "energy"  # later becomes dropdown in Streamlit

genre_avg = (
    genre_df
    .groupby("track_genre")[feature]
    .mean()
    .reset_index()
    .sort_values(by=feature, ascending=False)
)

fig = px.bar(
    genre_avg,
    x=feature,
    y="track_genre",
    orientation="h",
    title=f"Average {feature.capitalize()} by Genre",
    color=feature,
    template="plotly_dark"
)

fig.show()
```

```python
top_n = 10

top_genres = genre_counts.head(top_n)

fig = px.pie(
    top_genres,
    names="track_genre",
    values="count",
    title="Top Genres by Dataset Share",
    hole=0.4
)

fig.update_traces(textposition="inside", textinfo="percent+label")
fig.update_layout(template="plotly_dark")

fig.show()
```

```python
fig = px.box(
    genre_df,
    x="track_genre",
    y=feature,
    title=f"{feature.capitalize()} Distribution Across Genres",
    template="plotly_dark"
)

fig.show()
```

```python
chart_df = pd.read_csv(r"D:\spotify\data\processed\analytics_ready\chart_vs_nonchart.csv")
```

```python
chart_df.head()
```

```python
feature = "energy"  # will become dropdown later

mean_df = (
    chart_df
    .groupby("is_chart_song")[feature]
    .mean()
    .reset_index()
)

mean_df["is_chart_song"] = mean_df["is_chart_song"].map(
    {True: "Chart", False: "Non-Chart"}
)

fig = px.bar(
    mean_df,
    x="is_chart_song",
    y=feature,
    color="is_chart_song",
    title=f"Average {feature.capitalize()} — Chart vs Non-Chart Songs",
    template="plotly_dark",
    text_auto=".2f"
)

fig.show()
```

```python
features = [
    "danceability",
    "energy",
    "valence",
    "loudness",
    "tempo"
]

mean_features = (
    chart_df
    .groupby("is_chart_song")[features]
    .mean()
    .T
    .reset_index()
)

mean_features.columns = ["feature", "non_chart", "chart"]

mean_features["difference"] = mean_features["chart"] - mean_features["non_chart"]
mean_features = mean_features.sort_values("difference", ascending=False)

fig = px.bar(
    mean_features,
    x="difference",
    y="feature",
    orientation="h",
    title="Feature Differences: Chart vs Non-Chart Songs",
    template="plotly_dark",
    text_auto=".3f"
)

fig.show()
```

```python
fig = px.scatter(
    mean_features,
    x="difference",
    y="feature",
    size=mean_features["difference"].abs(),
    color="difference",
    title="Impact of Audio Features on Chart Presence",
    template="plotly_dark"
)

fig.show()
```

```python
fig = px.box(
    chart_df,
    x="is_chart_song",
    y=feature,
    title=f"{feature.capitalize()} Distribution: Chart vs Non-Chart",
    template="plotly_dark"
)

fig.show()
```

```python
pca_df = pd.read_csv(r"D:\spotify\data\processed\analytics_ready\audio_feature_pca.csv")
```

```python
pca_df.head()
```

```python
threshold = 70  # later becomes a slider in Streamlit

filtered_df = pca_df[pca_df["popularity"] >= threshold]

fig = px.scatter(
    filtered_df,
    x="PC1",
    y="PC2",
    color="popularity",
    title=f"PCA of Audio Features (Popularity ≥ {threshold})",
    color_continuous_scale="viridis",
    template="plotly_dark"
)

fig.show()
```

```python
fig = px.density_contour(
    pca_df,
    x="PC1",
    y="PC2",
    title="Density of Songs in Audio Feature Space",
    template="plotly_dark"
)

fig.show()
```

```python
fig = px.scatter(
    pca_df,
    x="PC1",
    y="PC2",
    color="popularity",
    opacity=0.4,
    title="PCA with Density Perspective",
    color_continuous_scale="viridis",
    template="plotly_dark"
)

fig.show()
```

```python
pop_cutoff = pca_df["popularity"].quantile(0.90)

top_hits = pca_df[pca_df["popularity"] >= pop_cutoff]

fig = px.scatter(
    pca_df,
    x="PC1",
    y="PC2",
    opacity=0.2,
    color_discrete_sequence=["gray"],
    title="Top 10% Popular Songs Highlighted",
    template="plotly_dark"
)

fig.add_scatter(
    x=top_hits["PC1"],
    y=top_hits["PC2"],
    mode="markers",
    marker=dict(color="red", size=6),
    name="Top 10% Popular Songs"
)

fig.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
pca_df["cluster"] = kmeans.fit_predict(pca_df[["PC1", "PC2"]])

fig = px.scatter(
    pca_df,
    x="PC1",
    y="PC2",
    color="cluster",
    title="KMeans Clusters in Audio Feature Space",
    template="plotly_dark"
)

fig.show()
```
